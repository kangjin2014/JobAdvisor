```python
import pandas as pd
import numpy as np
import os
import math
import docx2txt
from utils import load_skills_dict
from utils import match_kw
from utils import counter_cosine_similarity
import yaml
```

```python
with open("config.yml", 'r') as ymlfile: cfg = yaml.load(ymlfile)
creator = cfg['basic_info']['created_by']
print ("This Job Recommender is developed by {}".format(creator))
```

```python
my_text = docx2txt.process('/users/jinkang/job-recommender/' + cfg['file_path']['resume_uploaded']).lower()

smr_skills = load_skills_dict()

tmp_1 = match_kw(my_text,smr_skills)
tmp_2 = pd.read_csv('/users/jinkang/job-recommender/job-parser/file_saved/ds_2018-02-18.csv').\
            drop('Unnamed: 0', axis=1).\
            description.\
            apply(match_kw, args=(smr_skills,)).\
            fillna(value=0).\
            astype(int)
            
# has to define the apply in column-wise.
smlrt = tmp_2.apply(counter_cosine_similarity, args=(tmp_1,), axis = 1).\
              sort_values(ascending = False)
```

```python
import job_pst_indeed as jpi

df_jps = jpi.main('data scientist', 'north york', 'on')
```

```python
df_all_job = pd.DataFrame(df_jps, columns = ['title','company','location','date','description','url'])
df_recommend = df_all_job.loc[smlrt.index][0:10]
```
