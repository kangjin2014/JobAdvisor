```python
def create_new_column(a):
    
    # very useful
    d = ast.literal_eval(a)
    
    try:
        job_0 = d[0]
    except:
        job_0 = 'Null'
    try:
        job_1 = d[1]
    except:
        job_1 = 'Null'
    try:
        job_2 = d[2]
    except:
        job_2 = 'Null'
    try:
        job_3 = d[3]
    except:
        job_3 = 'Null'
        
    return job_0,job_1,job_2,job_3
```

```python
df_resume = pd.DataFrame(df['Unnamed: 0'].apply(create_new_column).tolist(), columns = ['job_0', 'job_1','job_2','job_3'])
df_resume
```
